**10.Define a function that takes a pandas DataFrame as an input and returns two pandas DataFrames: the first containing numerical continuous columns and the second containing numerical discrete columns of the input dataframe. To this end, it might be helpful to count the number of unique values. The function must have an optional argument set by default to 36 to discriminate between continuous and discrete columns. Then, use it to create two new dataframes: continuous_df and discrete_df.**

```python
def numericals_classifier(df):
    df1 = df.copy()
    num_df = df1.select_dtypes(np.number)
    continuous = []
    discrete = []
    for col in num_df.columns:
        if df[col].nunique() > 36:
             continuous.append(df[col])
        else:
            discrete.append(df[col])
    continuous_df = pd.DataFrame(continuous).T
    discrete_df = pd.DataFrame(discrete).T 
    return continuous_df, discrete_df
```

**1.Define a function that given a pandas DataFrame as input creates a seaborn countplot of each categorical column. Make sure to sort the bars by frequency ie: the most frequent values should be placed first. Hint: use .value_counts(). In addition, if the amount of unique values of a categorical column (cardinality) is six or more, the corresponding countplot should have the bars placed on the y-axis instead of the x-axis.**

```python
def countplot_generator(df):
    plt_size = len(df.columns)
    fig, ax = plt.subplots(plt_size, figsize=(8,50))
    n = 0
    for col in df.columns:
        if (len(df[col].value_counts().index)) >= 6:
            sns.countplot(data = df, y = col, order = df[col].value_counts().index, ax = ax[n])
        else:
            sns.countplot(data = df, x = col, order = df[col].value_counts().index, ax = ax[n])
        n +=1
    plt.tight_layout()      
    plt.show()
```

**13.Define a function to remove the outliers of a numerical continuous column depending if a value is bigger or smaller than a given amount of standard deviations of the mean (thr=3).**

```python
#import statistics as stats
def outlier_remover(df):
    df1 = df.copy()
    thr=3
    to_remove = []
    for col in df1.columns:
        sd_dw = np.mean(df1[col]) - (thr*(df1[col].std()))
        sd_up = np.mean(df1[col]) + (thr*(df1[col].std()))
        out = df1[(df1[col] < sd_dw)|(df1[col] > sd_up)]
        to_remove += list(out.index)
    df1 = df1.drop(to_remove)
    df1 = df1.reset_index(drop=True)
    return df1
```

Next, cast the resulting numpy arrays into pandas DataFrames. Make sure that the column names of the new dataframes are correctly setup using encoder.get_feature_names_out() and the same indexes of X_train_cat and X_test_cat**

**9.Create a function that given a model prediction and real values returns a pandas dataframe with the following table:**

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

def metric_tester(y, y_pred):
    data = [['MAE', mean_absolute_error(y, y_pred)], ['MAE', mean_squared_error(y, y_pred)], 
            ['RMSE', np.sqrt(mean_absolute_error(y, y_pred))], ['MAPE', mean_absolute_percentage_error(y, y_pred)],
           ['R2', r2_score(y, y_pred)]]
    df = pd.DataFrame(data, columns=['Error_metric', 'Value'])
    return df
    
```

**11.Now define a function that takes as an input: list of models, X_train and y_train to train several model (with default values) so we can train a lot of them without repeating code. The function must return the list of trained models.**

```python
def model_trainer(models_list, X_train, y_train, X_test):
    models_trained=[]
    for m in models_list:
        m.fit(X_train, y_train)
        y_train_pred = m.predict(X_train)
        y_test_pred = m.predict(X_test)
        models_trained.append((m, y_train_pred, y_test_pred))
        
    return models_trained
```

**12.Use the function to train the following model (with default settings): LinearRegressor, KNeighborsRegressor and MLPRegressor**
